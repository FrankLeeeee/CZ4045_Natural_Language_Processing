```python
import pandas as pd
import numpy as np
import os
import json
```

```python
json_directory = '.'

json_file = os.path.join(json_directory, 'reviewSelected100.json')

with open (json_file, 'rb') as f:
    data = f.read()

    
# 'utf-8' codec can't decode byte 0xe9 in position 30289: invalid continuation byte
# b'\xe9' - need latin-1 decoding

data = data.decode('latin-1').encode().decode()
```

```python
columns = list(json.loads(data.split('\n')[0]).keys())

rows = []
for r in data.split('\n')[:-1]:
    rows.append(list(json.loads(r).values()))

print(len(rows))
```

```python
df = pd.DataFrame(data = rows , columns= columns)
```
